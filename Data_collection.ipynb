```python
# web scraping it means extracting the data from net

import requests 
URL = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(URL)
```

```python
res
```

```python
if res.status_code == 200:
    print(res.content)
```

```python
if res.status_code == 200:
    print(res.headers)

    with open("scraped_data/data1.html", "w", encoding="utf-8") as f:
        f.write(res.text)
```

```python
# beautifulsap 

from bs4 import BeautifulSoup

with open("scraped_data/data1.html","r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content,"lxml")
```

```python
soup.find("h1")
```

```python
soup.find("h3")
```

```python
all_h3 = soup.find_all("h3")
for h3 in all_h3:
    print(h3)
```

```python
all_h3 = soup.find_all("h3")
for h3 in all_h3:
    print(h3.get_text()) # name of different country
```

```python
all_h3 = soup.find_all("h3")
for h3 in all_h3:
    name =h3.get_text(strip = True)
    print(h3.find_next("div"))
```

```python
all_h3 = soup.find_all("h3")
all_countries = []
for h3 in all_h3:
    name =h3.get_text(strip = True)

    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    print(h3.find_next("div").select_one("span.country-population").get_text(strip =True))
    all_countries.append([name,population])
```

```python
# countries and population
all_countries 
```

```python
# we can also restore as table using the pandas
import pandas as pd 
pd.DataFrame(all_countries)
```

```python
df = pd.DataFrame(all_countries,columns = ['Name','Population'])
df
```

```python
# convert this data into csv , then save then by creating a file
df.to_csv("cleaned_data/data.csv",index = False)
```
