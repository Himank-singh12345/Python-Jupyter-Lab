```python
# Melt - wide to long format

df = pd.DataFrame({
    "country" : ["USA","USA","India","India"],
    "year" :[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
    "tax" :[10,20,30,40]      
})

# wider number of column to longer number of row 
melted_df= df.melt(
    id_vars = ["country","year"],
    value_vars = ["sales","profit"],
    var_name ="metrics",
    value_name ="value"
)
```

```python
# pivot it longer to wider format
melted_df.pivot(
    index =["country","year"], # the which you want to keep
    columns ="metrics",
    values = "value"
)
```

```python
# merging & joinig  data

df_customers = pd.DataFrame({
    "customer_id":[1,2,3,4],
    "name" :["Adam" ,"Bob","Charlie","Dave"]
})

df_orders = pd.DataFrame({
    "order_id":[101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount" :[250,120,300,180]
})
```

```python
df_customers
```

```python
df_orders
```

```python
# merge 
# inner join = all common value has been merge together
pd.merge(df_customers,df_orders, on ="customer_id")
```

```python
# left join
pd.merge(df_customers,df_orders, on ="customer_id",how ="left")
```

```python
# right join
pd.merge(df_customers,df_orders, on ="customer_id",how = "right")
```

```python
# outer join
pd.merge(df_customers,df_orders, on ="customer_id", how = "outer")
```
