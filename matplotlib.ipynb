```python
# stack plots 
# here we use the pie chart for showing stack plots for only one days
# it is type of plot where multiple data series are stacked on top of each other
# each layer shows the contribution of one category and top line shows the cumulative total

days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
direct = [50,60,70,80,90,100,110]
organic = [30,40,50,55,60,70,80]
social = [20,25,30,35,40,50,60]
# we here use the autopct =%1.1% to shwo the percantage
# this is data come for single days only , so we use pie chart but , not for hole 7 days

plt.pie(
    [direct[0], organic[0], social[0]],
    labels=["Direct","Organic","Social"],
    autopct="%1.1f%%"
)
```

```python
# Stack plot
# Now we actually Using the Stack plot for showing the hole 7 days values because pie chart can not show the hole 7 days values thus the we use the stack plot now 
# Stack plot is type of plot where multiple data series are stacked on top of each other
# each layer shows the contribution of one category and top line shows the cumulative total

import matplotlib.pyplot as plt

days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]

direct = [50,60,70,80,90,100,110]
organic = [30,40,50,55,60,70,80]
social = [20,25,30,35,40,50,60]


# then to show for 7 days then we stack bar
plt.stackplot(days,direct,organic,social,labels=["Direct","Organic","Social"])
plt.title("Mraketing data for the week")
plt.xlabel("Days")
plt.ylabel("# of customers")
plt.legend()
plt.show()
```

```python
# Modern Matplotlib
# it is same as above but more advance modern which use for creating multple subplot in one image
# whenever we are working on indiviuable part the, we use the access obejct
# when we wanto works on over all canvas then we use the figure object
# this obejct oriented API use for multiple plot here
"""  
fig  Represents the entire figure

ax  Represents individual plots

subplots()  Creates multiple plots at once and returns fig & ax

ax[row][col]  Selects a specific plot location 

OR

fig = full page

ax = boxes on page

subplots() = best practice

ax[row][col] = graph position"""
# Creates 2 rows  2 columns = 4 plots (subplots(2,2))
# figsize = (width, height) , width=7,height=7
fig,ax = plt.subplots(2,2,figsize =(7,7)) 

# it first see [0] which first row seen plot in 0 position from row and column which show [0] position
ax[0][0].plot(x,y1)
# it use to write litle
ax[0][0].set_title("plot1 - square root")

# it [0] is first row position and then [1] is right side columns plot
ax[0][1].plot(x,y2)
# it use to write litle
ax[0][1].set_title("plot2 - double")

# ax[1][0] is [1] is row position but row is started from 0 and [0] is column position 
ax[1][0].plot(x,y3)
# it use to write litle
ax[1][0].set_title("plot3 - square ")

ax[1][1].plot(x,y4)
# it use to write litle
ax[1][1].set_title("plot4 - cube")

plt.tight_layout()
plt.suptitle("Multiple plots")
```

```python
# Temperature over cities
# compare temperature of cities over a week using subplots & loops

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

# Creates 2 rows  2 columns = 4 plots (subplots(2,2))
# this 2 - D array
fig,axes = plt.subplots(2,2) # ( 2 row and 2 columns need for (2+2 = 4 cities data()

city_no = 0
for i in range(2): # here rows (i)- 0,1
    for j in range(2): # here cols(j)- 0,1
       axes[i][j].plot(days,temperatures[city_no], marker="o") # marker is show the bold points
       axes[i][j].set_title(cities[city_no])
       axes[i][j].grid(True)
       city_no = city_no + 1 # here each time city is show the different cities

fig.suptitle("Temperature in cities over the week")

fig.tight_layout()
```

```python
# Temperature over cities
# compare temperature of cities over a week using subplots & loops

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

# Creates 2 rows  2 columns = 4 plots (subplots(2,2))

# it is now two D array
fig,axes = plt.subplots(2,2)

city_no = 0
for i in range(2): # here rows (i)- 0,1
    for j in range(2): # here cols(j)- 0,1
       axes[i][j].plot(days,temperatures[city_no], marker="o")
       axes[i][j].set_title(cities[city_no])
       axes[i][j].grid(True)
       city_no = city_no + 1

fig.suptitle("Temperature in cities over the week")
# it title as "Temperature" in y -axis
fig.supylabel("Temperature")
# it title as "Dyas" in x -axis
fig.supxlabel("Days")

fig.tight_layout()
```

```python
# Temperature over cities
# compare temperature of cities over a week using subplots & loops

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

# Creates 2 rows  2 columns = 4 plots (subplots(2,2))

# it is now 1 - D array
fig,axes = plt.subplots(2,2)

# this much simple way as compare to above code
# flatten use to convert the two array into one D-ARRAY
axes = axes.flatten()
for i,ax in enumerate(axes): # it show the index ,values at index , here "i" is use for (0,1,2,3) as index number , whereas "ax" is values
    ax.plot(days,temperatures[i],marker="o")
    ax.set_title(cities[i])
    ax.grid(True)
   
fig.suptitle("Temperature in cities over the week")
# it title as "Temperature" in y -axis
fig.supylabel("Temperature")
# it title as "Dyas" in x -axis
fig.supxlabel("Days")

fig.tight_layout()
```
